## word frequency using dictionary

```python
m='one two three four five one two five six seven one eight four five'
```

```python
m1=m.split()
m1
```

```python
c=[0]*len(m2)
my=dict()
for i in range(len(m2)):
    for j in range(len(m1)):
        if m2[i]==m1[j]:
            c[i]+=1
    my[m2[i]]=c[i]
print(my)
```
